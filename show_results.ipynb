```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_orig = pd.read_csv('test-data2.csv', )
df_orig = df_orig.set_index('ID')
df_sel = pd.read_csv('selected.csv')
df_sel = df_sel.set_index('ID')
```

```python
fig, ax = plt.subplots()
ax.scatter(df_orig['X1'], df_orig['X2'], color='b')
ax.scatter(df_sel['X1'], df_sel['X2'], color='y', marker='x')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(df_orig['X1'], df_orig['X3'], color='b')
ax.scatter(df_sel['X1'], df_sel['X3'], color='y', marker='x')
plt.show()
```
