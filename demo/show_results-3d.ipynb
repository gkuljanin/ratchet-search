```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_orig = pd.read_csv('test-data2.csv')
df_orig = df_orig.set_index('ID')
df_sel = pd.read_csv('selected.csv')
df_sel = df_sel.set_index('ID')
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')
ax.scatter(df_orig['X1'], df_orig['X2'], df_orig['X3'], color='b')
ax.scatter(df_sel['X1'], df_sel['X2'], df_sel['X3'], color='r', marker='x', s=100)
ax.invert_xaxis()
for x, y, z, label in zip(df_orig['X1'], df_orig['X2'], df_orig['X3'], df_orig.index):
    ax.text(x, y, z, label)
```
