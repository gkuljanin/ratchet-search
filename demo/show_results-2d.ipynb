```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_orig = pd.read_csv('test-data2.csv')
df_orig = df_orig.set_index('ID')
df_sel = pd.read_csv('selected-2d-1.csv')
df_sel = df_sel.set_index('ID')
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot()
ax.scatter(df_orig['X1'], df_orig['X2'], color='b')
ax.scatter(df_sel['X1'], df_sel['X2'], color='r', marker='x', s=100)
for x, y, label in zip(df_orig['X1'], df_orig['X2'], df_orig.index):
    ax.text(x, y, label)
```

```python
df_sel2 = pd.read_csv('selected-2d-2.csv')
df_sel2 = df_sel2.set_index('ID')
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot()
ax.scatter(df_orig['X1'], df_orig['X2'], color='b')
ax.scatter(df_sel2['X1'], df_sel2['X2'], color='r', marker='x', s=100)
for x, y, label in zip(df_orig['X1'], df_orig['X2'], df_orig.index):
    ax.text(x, y, label)
```
